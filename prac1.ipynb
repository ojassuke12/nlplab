```python
import nltk
from nltk.corpus import genesis
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import pos_tag

nltk.download('all')
# Download the required NLTK corpora and models
nltk.download('genesis')
nltk.download('punkt')
nltk.download('stopwords')

# Load the Genesis corpus and extract the text
genesis_text = genesis.raw()

# Split the text into a list of words
words = word_tokenize(genesis_text)

# Remove punctuation and stopwords
punctuation = [',', '.', ';', ':', '!', '?', '-', '(', ')', '[', ']', '{', '}']
stop_words = set(stopwords.words('english'))
filtered_words = [word.lower() for word in words if word not in punctuation and word.lower() not in stop_words]

# Perform stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

# Perform POS tagging
tagged_words = pos_tag(stemmed_words)

# Print the tagged words
print(tagged_words)
```
