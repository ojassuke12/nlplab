```python
import random
import nltk

# Download the names corpus if necessary
nltk.download('names')

# Get the male and female names from the corpus
male_names = nltk.corpus.names.words('male.txt')
female_names = nltk.corpus.names.words('female.txt')

# Combine the male and female names
all_names = male_names + female_names

# Shuffle the list of names
random.shuffle(all_names)

# Print the first 15 combined labeled names
for name in all_names[:15]:
    if name in male_names:
        print(f"male\t{name}")
    else:
        print(f"female\t{name}")

```
